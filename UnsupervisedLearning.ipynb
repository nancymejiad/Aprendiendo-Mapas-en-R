# Introducción: Aprendizaje No Supervisado

El aprendizaje no supervisado es una rama fundamental del aprendizaje automático que se enfoca en **analizar datos sin etiquetas, ni valores objetivo**.

A diferencia del aprendizaje supervisado, donde los modelos aprenden a partir de ejemplos con salidas conocidas, en el aprendizaje no supervisado, los algoritmos deben descubrir por sí mismo patrones, estructuras o relaciones ocultas en los datos.

Esta capacidad los convierte en una herramienta poderosa para explorar información en bruto, especialmente en contextos donde etiquetar los datos resulta costoso o inviable.

## Objetivos


Uno de los principales objetivos del aprendizaje no supervisado es la **organización y comprensión de grandes volúmenes de datos**.

Al no depender de supervisión humana directa, estos algoritmos pueden:

* **Identificar grupos naturales**
* **Reducir la complejidad de los datos**
* **Detectar comportamientos atípicos**

Esto lo hace especialmente útil en áreas como la segmentación de clientes, la bioinformática, la detección de fraudes, el análisis de redes sociales y la compresión de información.



## Extensión al aprendizaje profundo

El aprendizaje no supervisado también se ha integrado con técnicas de aprendizaje profundo, dando lugar a modelos más sofisticados como:

* **Autoencoders**
* **Redes generativas adversarias (GANs)**
* **Aprendizaje contrastivo**

Estos modelos permiten representar los datos en espacios latentes más ricos, generar nuevas muestras sintéticas o aprender representaciones útiles para tareas posteriores.

En conjunto, el aprendizaje no supervisado constituye una base esencial para el análisis exploratorio y la comprensión automatizada de datos complejos.

## Métricas para Agrupamiento

<img src="https://drive.google.com/uc?export=view&id=1xq3W4STA-zxD1MKs3t-GW86xCNiWTIcT" alt="objetivos visualizacion" width="500"/>

### 1. **Índice de Calinski-Harabasz (CH)**

- Mide la conexión entre la dispersión de los grupos (BSS inter-cluster) y la dispersión dentro de los grupos (WSS intra-cluster).

$$
CH = \frac{\text{BSS} / (k - 1)}{\text{WSS} / (n - k)}, \quad
\text{BSS} = \sum_{i=1}^{k} n_i \| c_i - c \|^2, \quad
\text{WSS} = \sum_{i=1}^{k} \sum_{x \in C_i} \| x - c_i \|^2
$$

- **Mayor valor = mejor agrupamiento**: Indica que los grupos están bien separados y son internamente compactos.
- No tiene un límite máximo.
- Uso: Selección del número óptimo de *k* cuando los datos tienen estructura clara.

### 2. **Índice de Davies-Bouldin (DBI)**

- Mide el promedio de la similitud entre cada grupo y su grupo más cercano, considerando dispersión y distancia.

$$
\text{DBI} = \frac{1}{k} \sum_{i=1}^{k} \max_{j \ne i} \left( \frac{S_i + S_j}{M_{ij}} \right), \quad
S_i = \frac{1}{|C_i|} \sum_{x \in C_i} \| x - c_i \|, \quad
M_{ij} = \| c_i - c_j \|
$$

- **Menor valor = mejor agrupamiento**, un valor cercano a cero indica mejor separación y compacidad de los grupos.
- Penaliza grupos que se solapan o que son muy dispersos, que se observa con un valor alto, pero no hay un máximo teórico.
- Es sensible a la forma y densidad de los grupos.

### 3. **Índice de Silhouette**

- Mide qué tan similar es un punto a su propio grupo comparado con otros grupos.

$$
\text{Silhouette Score} = \frac{1}{N} \sum_{i=1}^{N} s(i), \quad s(i) = \frac{b(i) - a(i)}{\max\{a(i), b(i)\}}
$$

$$
 a(i) = \frac{1}{|C_i| - 1} \sum_{x \in C_i, x \ne i} \| x - i \|, \quad b(i) = \min_{j \ne i} \left( \frac{1}{|C_j|} \sum_{x \in C_j} \| x - i \| \right) $$

$$ \text{mal asignado } (-1)  \leq s(i)= \text{Puntos en frontera } (0) \leq \text{Bien asignado }(1)$$

- Uso: Visualizar la estructura interna y validar la cohesión/separación de los grupos.

### Comparación resumida

| Índice              | Ideal | Indica...                          |
|---------------------|--------|------------------------------------|
| Calinski-Harabasz   | ↑      | Separación clara entre grupos   |
| Davies-Bouldin      | ↓      | Baja superposición entre grupos    |
| Silhouette          | ↑      | Buena asignación y cohesión        |

```python
from sklearn.metrics import classification_report
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import plotly.graph_objects as go
from sklearn.cluster import KMeans, AgglomerativeClustering, DBSCAN

# Reducción de dimensión
def reducir_dim(X, metodo='pca', n_componentes=3, **kwargs):
    if metodo == 'pca':
        modelo = PCA(n_components=n_componentes, **kwargs)
    elif metodo == 'umap':
        modelo = umap.UMAP(n_components=n_componentes, **kwargs)
    elif metodo == 'kpca':
        modelo = KernelPCA(n_components=n_componentes, **kwargs)
    elif metodo == 'isomap':
        modelo = Isomap(n_components=n_componentes, **kwargs)
    else:
        raise ValueError(f"Método '{metodo}' no reconocido.")

    X_reducido = modelo.fit_transform(X)
    return X_reducido, modelo

def construir_figura(frames):
    fig = go.Figure(
        data=frames[0].data,
        layout=go.Layout(
            title='Iteración 0',
            scene=dict(xaxis_title='Comp1', yaxis_title='Comp2', zaxis_title='Comp3'),
            sliders=[dict(
                active=0,
                steps=[dict(
                    method='animate',
                    args=[[f.name], {
                        "mode": "immediate",
                        "frame": {"duration": 0, "redraw": True},
                        "transition": {"duration": 0}
                    }],
                    label=f"Iteración {f.name}"
                ) for f in frames],
                x=0.1, y=0, len=0.9
            )]
        ),
        frames=frames
    )
    fig.show()

def simular_clustering_2d(X, y_real, modelo_reduccion, algoritmo='kmeans', n_clusters=3, n_iteraciones=10):
    np.random.seed()
    frames = []
    symbol_map = {0: 'square', 1: 'cross', 2: 'circle'}
    color_map = {0: 'red', 1: 'green', 2: 'orange'}

    if modelo_reduccion == None:
        X_red = X
    else:
        X_red = modelo_reduccion.transform(X)

    if algoritmo.lower() == 'kmeans':
        # Inicialización aleatoria de centroides
        centroides = X[np.random.choice(len(X), n_clusters, replace=False)]
        for iteracion in range(n_iteraciones):
            distancias = np.linalg.norm(X[:, np.newaxis] - centroides, axis=2)
            etiquetas = np.argmin(distancias, axis=1)
            centroides = np.array([X[etiquetas == i].mean(axis=0) for i in range(n_clusters)])
            if modelo_reduccion == None:
                centroides_red = centroides
            else:
                centroides_red = modelo_reduccion.transform(centroides)

            trazas = []
            for etiqueta in np.unique(y_real):
                idx = np.where(y_real == etiqueta)[0]
                trazas.append(go.Scatter(
                    x=X_red[idx, 0],
                    y=X_red[idx, 1],
                    mode='markers',
                    marker=dict(
                        size=6,
                        symbol=symbol_map.get(etiqueta, 'circle'),
                        color=[color_map.get(etiquetas[i], 'gray') for i in idx]
                    ),
                    name=f"Clase real {etiqueta}",
                    showlegend=(iteracion == 0)
                ))

            trazas.append(go.Scatter(
                x=centroides_red[:, 0],
                y=centroides_red[:, 1],
                mode='markers',
                marker=dict(size=8, color='blue', symbol='x'),
                name='Centroides'
            ))

            frames.append(go.Frame(data=trazas, name=str(iteracion)))

    elif algoritmo.lower() == 'agglomerative':
        etiquetas = AgglomerativeClustering(n_clusters=n_clusters).fit_predict(X)
        trazas = []
        for etiqueta in np.unique(y_real):
            idx = np.where(y_real == etiqueta)[0]
            trazas.append(go.Scatter(
                x=X_red[idx, 0],
                y=X_red[idx, 1],
                mode='markers',
                marker=dict(
                    size=6,
                    symbol=symbol_map.get(etiqueta, 'circle'),
                    color=[color_map.get(etiquetas[i], 'gray') for i in idx]
                ),
                name=f"Clase real {etiqueta}"
            ))
        frames.append(go.Frame(data=trazas, name="0"))

    elif algoritmo.lower() == 'dbscan':
        etiquetas = DBSCAN(eps=0.5, min_samples=5).fit_predict(X)
        trazas = []
        for etiqueta in np.unique(y_real):
            idx = np.where(y_real == etiqueta)[0]
            trazas.append(go.Scatter(
                x=X_red[idx, 0],
                y=X_red[idx, 1],
                mode='markers',
                marker=dict(
                    size=6,
                    symbol=symbol_map.get(etiqueta, 'circle'),
                    color=[color_map.get(etiquetas[i], 'gray') for i in idx]
                ),
                name=f"Clase real {etiqueta}"
            ))
        frames.append(go.Frame(data=trazas, name="0"))

    return frames

# Mapeo óptimo de grupos a etiquetas reales (algoritmo húngaro)
from sklearn.metrics import confusion_matrix
from scipy.optimize import linear_sum_assignment
import numpy as np

def mapear_clusters_hungaro(y_true, y_pred):
    y_true = np.array(y_true)
    y_pred = np.array(y_pred)

    # Clases y clusters únicos
    clases = np.unique(y_true)
    clusters = np.unique(y_pred)

    # Matriz de confusión con orden explícito
    matriz_conf = confusion_matrix(y_true, y_pred, labels=clases)

    # Aplicar algoritmo húngaro sobre la matriz negativa
    fila_ind, col_ind = linear_sum_assignment(-matriz_conf)

    # Construir mapeo: cluster → clase
    mapeo = {clusters[col]: clases[fila] for fila, col in zip(fila_ind, col_ind)}

    # Aplicar mapeo
    y_mapeado = np.array([mapeo.get(c, c) for c in y_pred])

    return y_mapeado, mapeo, matriz_conf

def mapear_clusters_manual(target, y_pred, mapeo_manual=None, etiquetas_nominales=None):
    target = np.array(target)
    y_pred = np.array(y_pred)

    # Clusters únicos (incluyendo ruido)
    clusters = sorted(set(y_pred))
    clases_reales = sorted(set(target))

    # Inicializar mapeo si no se da
    if mapeo_manual is None:
        mapeo_manual = {}

    # Construir mapeo extendido
    mapeo_completo = {}
    etiquetas_final = []

    # Contador para clases extra desde cero (evita solapar clases reales)
    clase_extra_base = max(clases_reales) + 1
    clase_extra_actual = clase_extra_base

    for cluster in clusters:
        if cluster == -1:
            mapeo_completo[-1] = -1  # Ruido
        elif cluster in mapeo_manual:
            mapeo_completo[cluster] = mapeo_manual[cluster]
        else:
            mapeo_completo[cluster] = clase_extra_actual
            clase_extra_actual += 1

    # Mapear y_pred
    y_mapeado = np.array([
        mapeo_completo.get(c, -1) for c in y_pred
    ])

    # Construir etiquetas nominales
    clases_mapeadas = sorted(set(y_mapeado))
    for clase in clases_mapeadas:
        if clase == -1:
            etiquetas_final.append("Ruido")
        elif etiquetas_nominales and clase in etiquetas_nominales:
            etiquetas_final.append(etiquetas_nominales[clase])
        elif clase in clases_reales:
            etiquetas_final.append(f"Clase {clase}")
        else:
            etiquetas_final.append(f"Extra {clase}")

    return y_mapeado, etiquetas_final

def graficar_silhouette_clusters(modelo, X):
    # Obtener etiquetas
    y_pred = modelo.labels_ if hasattr(modelo, "labels_") else modelo.predict(X)

    # Filtrar puntos de ruido (-1) si existen
    mask = y_pred != -1
    X_filtrado = X[mask]
    y_filtrado = y_pred[mask]

    # Verificar si hay al menos 2 clusters para calcular silhouette
    if len(np.unique(y_filtrado)) < 2:
        print("No hay suficientes clusters para calcular Silhouette.")
        return

    # Calcular silhouette
    sil_vals = silhouette_samples(X_filtrado, y_filtrado)
    sil_avg = silhouette_score(X_filtrado, y_filtrado)

    # Graficar
    plt.figure(figsize=(10, 6))
    y_lower = 10
    for i, cluster in enumerate(np.unique(y_filtrado)):
        sil_cluster = sil_vals[y_filtrado == cluster]
        sil_cluster.sort()
        y_upper = y_lower + len(sil_cluster)
        plt.fill_betweenx(np.arange(y_lower, y_upper),
                          0, sil_cluster, alpha=0.7)
        plt.text(-0.05, y_lower + 0.5 * len(sil_cluster), str(cluster))
        y_lower = y_upper + 10

    plt.axvline(sil_avg, color="red", linestyle="--")
    plt.xlabel("Silhouette coefficient")
    plt.ylabel("Cluster")
    plt.title(f"Silhouette plot (avg={sil_avg:.2f})")
    plt.show()

# Gráfica tres métricas de agrupamiento para divesos números de grupos
def graficar_indices_agrupados(ch_scores, dbi_scores, silhouette_scores):
    assert len(ch_scores) == len(dbi_scores) == len(silhouette_scores), "Las listas deben tener la misma longitud"

    k_vals = np.arange(2, len(ch_scores) + 2)
    ancho = 0.25  # ancho de cada barra

    fig, ax1 = plt.subplots(figsize=(9, 6))

    # Eje 1: Calinski-Harabasz
    barras_ch = ax1.bar(k_vals - ancho, ch_scores, width=ancho, color='darkorange', label='Calinski-Harabasz')
    ax1.set_ylabel('CH Index (↑ mejor)', color='darkorange')
    ax1.tick_params(axis='y', labelcolor='darkorange')
    ax1.grid(False)  # quita grid horizontal de este eje

    # Eje 2: Davies-Bouldin
    ax2 = ax1.twinx()
    barras_dbi = ax2.bar(k_vals, dbi_scores, width=ancho, color='steelblue', label='Davies-Bouldin')
    ax2.set_ylabel('DBI (↓ mejor)', color='steelblue')
    ax2.tick_params(axis='y', labelcolor='steelblue')
    ax2.grid(False)  # quita grid horizontal de este eje

    # Eje 3: Silhouette
    ax3 = ax1.twinx()
    ax3.spines["right"].set_position(("axes", 1.1))  # desplaza el tercer eje
    barras_sil = ax3.bar(k_vals + ancho, silhouette_scores, width=ancho, color='forestgreen', label='Silhouette')
    ax3.set_ylabel('Silhouette (↑ mejor)', color='forestgreen')
    ax3.tick_params(axis='y', labelcolor='forestgreen')
    ax3.grid(False)  # quita grid horizontal de este eje
    # Eje X
    ax1.set_xticks(k_vals)
    ax1.set_xticklabels([str(k) for k in k_vals])
    ax1.set_xlabel('Número de grupos (k)')
    ax1.set_title('Comparación de Índices para datos no supervisados')

    # Leyenda combinada
    handles = [barras_ch, barras_dbi, barras_sil]
    labels = ['Calinski-Harabasz', 'Davies-Bouldin', 'Silhouette']
    ax1.legend(handles, labels, loc='upper center', bbox_to_anchor=(0.5, -0.1), ncol=3)
    plt.grid(False)
    plt.tight_layout()
    plt.show()

# Visualización de matriz de confusión
def graficar_matriz_confusion(y_true, y_pred, etiquetas=None):
    # Detectar puntos de ruido (-1) y filtrarlos
    mask = y_pred != -1
    y_true_filtrado = y_true[mask]
    y_pred_filtrado = y_pred[mask]

    matriz = confusion_matrix(y_true_filtrado, y_pred_filtrado)

    # Si no se pasan etiquetas, usar los valores únicos de y_true_filtrado
    if etiquetas is None:
        etiquetas = np.unique(y_true_filtrado)

    plt.figure(figsize=(9, 6))
    sns.heatmap(matriz, annot=True, fmt='d', cmap='Blues',
                xticklabels=etiquetas, yticklabels=etiquetas)
    plt.title('Matriz de Confusión')
    plt.xlabel('Predicción')
    plt.ylabel('Etiqueta real')
    plt.tight_layout()
    plt.show()

# Mostrar classification_report con soporte automático para ruido (-1 en DBSCAN)
def mostrar_reporte_clasificacion(y_true, y_pred, etiquetas=None):
    clases_pred = np.unique(y_pred)
    clases_true = np.unique(y_true)

    # Detectar si hay más clases en y_pred que en y_true
    clases_extra = set(clases_pred) - set(clases_true)

    # Construir etiquetas automáticas si no se proporcionan
    if etiquetas is None:
        etiquetas_final = [f"Clase {i}" for i in sorted(clases_true)]
    else:
        etiquetas_final = list(etiquetas)

    # Agregar etiquetas para clases adicionales en y_pred
    for clase in sorted(clases_extra):
        if clase == -1:
            etiquetas_final.append("Ruido")
        else:
            etiquetas_final.append(f"Extra {clase}")

    # Generar reporte
    reporte = classification_report(
        y_true,
        y_pred,
        labels=sorted(clases_true.tolist() + list(clases_extra)),
        target_names=etiquetas_final,
        output_dict=True,
        zero_division=0  # Evita warnings por clases sin soporte
    )

    # Mostrar reporte textual
    print(classification_report(
        y_true,
        y_pred,
        labels=sorted(clases_true.tolist() + list(clases_extra)),
        target_names=etiquetas_final,
        zero_division=0
    ))

    # Convertir a DataFrame
    df = pd.DataFrame(reporte).transpose()

    # Filtrar métricas por clase (excluye 'accuracy', 'macro avg', etc.)
    df_clases = df.loc[etiquetas_final, ['precision', 'recall', 'f1-score']]

    # Graficar
    plt.figure(figsize=(9, 6))
    df_clases.plot(kind='bar', colormap='viridis')
    plt.title('Métricas por Clase')
    plt.ylabel('Valor')
    plt.ylim(0, 1.05)
    plt.xticks(rotation=45)
    plt.grid(axis='y', linestyle='--', alpha=0.5)
    plt.tight_layout()
    plt.show()
```

# K-means

El algoritmo de [K-means (MacQueen, 1967)](https://projecteuclid.org/download/pdf_1/euclid.bsmsp/1200512992) particiona el rango multidimensional del conjunto de entrada, definido por la métrica de distancia o similitud.  

- El objetivo es agrupar objetos en $k$ grupos basandose en sus características, minimizando la suma de las similitudes entre cada objeto y el centroide de su grupo o clúster.

- Es un algoritmo iterativo que puede concluir con diversos criterios de paro.

- Tiene una complejidad $O(n * k * i * d)$ donde $n$ es el número de datos, $k$ el número de grupos, $i$ el número de iteraciones y $d$ el número de atributos o dimensiones.



## Pseudocódigo

**Entrada**. Un conjunto de instancias $X = \{x_1, x_2, \dots, x_n\} \subset \mathbb{R}^d$, donde cada $x_i$ es un vector de dimensión $d$.

1. **Inicialización**. Proporcionar el número de grupos $k$, con este inicializar los centroides $M = \{m_1, m_2, \dots, m_k\} \subset \mathbb{R}^d$.  
La inicialización puede ser: Aleatoria, basada en heurísticas (ej. K-means++), determinística (ej. primeros $k$ puntos).

2. **Asignación de instancias a los centroides**. Para cada instancia $x_i \in X$, se asigna al centroide más cercano según una métrica de similitud:

$$
C_j ← x_i = \arg\min_{j \in \{1, \dots, k\}} dist( x_i - m_j)
$$

3. **Actualización de centroides**. Para cada grupo $C_j$, se actualiza el centroide como el promedio de los puntos asignados:

$$
m_j = \frac{1}{|C_j|} \sum_{x_i \in C_j} x_i
$$

4. **Condición de paro**. Repetir los pasos 2 y 3 hasta que se cumpla algunas condiciones, ej.  los centroides no cambian, el número de iteraciones alcanza un máximo predefinido o el número de puntos que cambian de grupo es menor que un umbral.

## Inicializar centroides

Una de las mejores maneras de inicializar los valores de los centroides es empleando el algoritmo modificado descrito en [**K-means++**](https://theory.stanford.edu/~sergei/papers/kMeansPP-soda.pdf). Su objetivo es reducir la probabilidad de convergencia hacia soluciones subóptimas, mejorando la estabilidad y calidad del agrupamiento.

### Pseudo-código

Entrada: Dado el conjunto de instancias $X$ y el número de grupos $k$.

1. Seleccionar aleatoriamente el primer centroide $m_1$ de $X$
2. Para cada punto $x$ en $X$, calcular la distancia mínima al centroide más cercano de los ya seleccionados. Seleccionar el siguiente centroide $m_i$ con probabilidad proporcional a $D(x)^2$, donde $D(x)$ es la distancia mínima de $x$ a cualquier centroide ya elegido.
3. Repetir el paso 1–2 hasta obtener $k$ centroides iniciales.

# Agrupamiento Jerárquico Aglomerativo
El [agrupamiento jerárquico aglomerativo](https://arxiv.org/pdf/1109.2378) (bottom-up o agglomerative hierarchical clustering) es un método de agrupamiento que construye una jerarquía de grupos mediante la fusión progresiva de instancias o grupos, comenzando con cada punto como un grupo individual en el nivel más bajo.


- Cada instancia del conjunto de datos se considera un clúster independiente.
- Se trabaja sobre una matriz de similitudes entre todos los clústeres (inicialmente entre puntos individuales).

- En cada paso, se fusionan los dos clústeres más cercanos según un criterio de enlace (linkage).

- Se actualiza (reduce) la matriz de distancias para reflejar la nueva estructura.

- Al final o después de un nivel deseado se representa un árbol jerárquico. El resultado es un dendrograma que muestra cómo se agrupan los datos a diferentes niveles de similitud.

## Criterios de enlace (linkage)

El criterio de enlace define cómo se fusionan los grupos por pares $A \cup B \rightarrow U$:

<img src="https://drive.google.com/uc?export=view&id=1gtkZtO2MTSXWm_sEKcg7EhvxbtVzJQTY" alt="objetivos visualizacion" width="1800"/>

- **Enlace simple** (single linkage) $ d(U,X) = \min(d(A,X),d(B,X)) \forall X \neq A \land \neq B$ Favorece cadenas al unir puntos cercanos, sensible al ruido.

- **Enlace completo** (complete linkage) $ d(U,X) = \max(d(A,X),d(B,X)) \forall X \neq A \land \neq B$ Produce clústeres compactos, evita fusiones con puntos alejados.

- **Enlace promedio** (average linkage)  $ d(U,X) = \frac{|A|d(A,X)+|B|d(B,X)}{|A|+|B|}$ Balance entre los anteriores, útil para estructuras suaves.

- **Enlace centroidal** (centroid linkage) $ d(U,X) = \| \mu_U - \mu_X \| $ Puede generar fusiones no monotónicas (efecto inverso), $\mu$ es el centroide de las instancias perteneciente a ese grupo.

- **Minima varianza** (Ward) $ d(U,X) = \frac{|U| \cdot |X|}{|U| + |X|} \, \| \mathbf{\mu}_U - \mathbf{\mu}_X \|^2 $ Minimiza la varianza intra-clúster (SSE) al fusionar.




## Ventajas

- No requiere especificar el número de clústeres a priori  
- Produce una estructura jerárquica útil para análisis exploratorio y visualmente ubicar clústers
- Compatible con diferentes métricas de similitud

## Limitaciones

- Escalabilidad limitada: complejidad computacional $ O(n^2) $  
- Sensible al ruido y a la escala de los datos  
- Las decisiones de fusión no se pueden revertir

## Pseudocódigo
**Entrada**. Un conjunto de instancias $X = \{x_1, x_2, \dots, x_n\} \subset \mathbb{R}^d$, donde cada $x_i$ es un vector de dimensión $d$.

1. **Inicializar**. Establecer nivel $l = 0$ y los clusters individuales $\mathcal{C}^{(l)} = \{ x_1, x_2, \dots, x_ \}$.

2. **Matriz de distancias**. $\{D_{ij} = \| x_i - x_j \|, i,j $: se calcula la matriz de distancias entre todos los pares de instancias.

3. **Distancia mínima**. Buscar el par de clústers $(A, B)$ más cercanos con la menor distancia en la matriz de distancias, aumentar nivel $l=l+1$.

4. **Fusión**: Unir los indices de las instancias de los clústers $A \cup B \rightarrow U$, el nuevo nivel es $\mathcal{C}^{(l)} = \{ \mathcal{C}^{(l-1)} \setminus \{A, B\} \} \cup \{U\}
$. y calcular las distancias del nuevo grupo en base al criterio del enlace $d(U,:)$.

5. **Actualizar la matriz de distancias** $D^{l} \in \mathbb{R}^{(n - l) \times (n - l)}$, eliminado renglón/columna de $A$ y $B$ y actualizar el renglón $U$ con $d(U,:)$.

6. Repetir 3-5 hasta tener un solo cluster o un número solicitado.

6. Repetir



## **Dendrograma**

Cada fusión se representa como un nodo en un árbol binario.
El resultado es un **dendrograma** que muestra la jerarquía de agrupamientos. El dendrograma nos permite visualizar información sobre los posibles grupos que se describen de manera natural.

Para obtener una partición final en $k$ grupos, se realiza un corte horizontal en el dendrograma a una altura determinada (distancia de fusión), para obtener las instancias por grupo se accede a través de $\mathcal{C^{n-k}}$.

# DBSCAN

DBSCAN (Density-Based Spatial Clustering of Applications with Noise) es un algoritmo de agrupamiento no paramétrico propuesto por [Martin Ester, Hans-Peter Kriegel, Jörg Sander y Xiaowei Xu en 1996](https://www2.cs.uh.edu/~ceick/7363/Papers/dbscan.pdf).

Se basa en la noción de densidad para identificar grupos de puntos en un espacio, diferenciando entre regiones densamente pobladas y áreas de baja densidad que se consideran ruido. Su principal fortaleza es la capacidad de detectar **clusters de forma arbitraria** y manejar datos con ruido sin necesidad de especificar el número de clusters de antemano.

**Conceptos clave**  
- **Punto núcleo:** Un punto con al menos $minPts$ vecinos dentro de un radio $\varepsilon$ (epsilon).  
- **Punto alcanzable directamente:** Un punto que está dentro de la distancia $\varepsilon$ de un punto núcleo.  
- **Punto alcanzable:** Existe una cadena de puntos donde cada uno es alcanzable directamente desde el anterior, comenzando en un punto núcleo.  
- **Ruido:** Puntos que no cumplen las condiciones para ser núcleo ni alcanzables desde un núcleo.

**Ventajas**  
- Detecta clusters de forma arbitraria.  
- No requiere conocer el número de clusters previamente.  
- Robusto frente a ruido y valores atípicos.  

**Limitaciones**  
- Sensible a la elección de $\varepsilon$ y $minPts$.  
- Puede tener dificultades con clusters de densidades muy variables.  
- Sensible a la escala de los datos


## Pseudocódigo

**Objetivo**. Agrupar puntos densamente conectados y detectar ruido, sin necesidad de especificar el número de clústeres.

**Entrada**. Un conjunto de instancias $X = \{x_1, x_2, \dots, x_n\} \subset \mathbb{R}^d$, donde cada $x_i$ es un vector de dimensión $d$.

1. **Inicialización**.
- Solicitar $ \varepsilon $: radio de vecindad (epsilon) y $ \text{minPts} $: número mínimo de puntos requeridos para formar una región densa.
- Marcar todos los puntos como **no visitados**.
- $l = 0$ grupos actuales.

2. Para cada punto $x_i$ no visitado:
   - Marcar como **visitado**
   - Obtener sus vecinos $N_\varepsilon(x_i)=\{x_j \in X : \|x_i - x_j\| \leq \varepsilon\}$
   - Si no tiene el mínimo de puntos etiquetar $x_i$ como **ruido** $P(i)=-1 \leftarrow |N_\varepsilon(x_i)| < \text{minPts}$
   - Entonces $x_i$ es **punto núcleo**:
     - Crear nuevo clúster $l=l+1$, $C_l$
     - Etiquetar $x_i$ como $P(i)=l$
     - Visitar sus vecinos: Para cada $x_j \in N_\varepsilon(x_i)$:
         - Si $x_j$ no visitado:
           - Marcar como visitado
           - Obtener $N_\varepsilon(x_j)$
           - Si $|N_\varepsilon(x_j)| \geq \text{minPts} \rightarrow N_\varepsilon(x_i) \cup N_\varepsilon(x_j)$, agregar vecinos a $N_\varepsilon(x_i)$
         - Si $x_j$ no pertenece a ningún clúster ($P(j)= NaN$) $\rightarrow$ agregar a $C_l$ y etiquetar $P(j) = l$

3. Repetir hasta visitar todos los puntos

4. **Salida**. Etiquetas $P$, que contiene los indices de cluster y ruido.

```python
def evaluar_dbscan_por_rango(data, eps_vals, min_samples_vals):
    resultados = []

    for eps in eps_vals:
        for min_samples in min_samples_vals:
            db = DBSCAN(eps=eps, min_samples=min_samples).fit(data)
            labels = db.labels_

            # Número de clusters reales (excluyendo ruido)
            n_clusters = len(set(labels)) - (1 if -1 in labels else 0)

            # Si todos los puntos son ruido o hay solo un cluster, no guardar
            if n_clusters <= 1:
                continue

            ch = calinski_harabasz_score(data, labels)
            dbi = davies_bouldin_score(data, labels)
            silhouette = silhouette_score(data, labels)

            resultados.append({
                "eps": eps,
                "min_samples": min_samples,
                "n_clusters": n_clusters,
                "ch_scores": ch,
                "dbi_scores": dbi,
                "silhouette_scores": silhouette
            })
    return resultados

# Ejemplo de rangos a evaluar
eps_vals = np.linspace(0.2, 5, 10)       # 10 valores de eps
min_samples_vals = [3, 5, 7, 9, 11]               # varios valores de min_samples

indices = evaluar_dbscan_por_rango(data, eps_vals, min_samples_vals)
ch_scores = [r["ch_scores"] for r in indices]
dbi_scores = [r["dbi_scores"] for r in indices]
silhouette_scores = [r["silhouette_scores"] for r in indices]

graficar_indices_agrupados(ch_scores, dbi_scores, silhouette_scores)
```
